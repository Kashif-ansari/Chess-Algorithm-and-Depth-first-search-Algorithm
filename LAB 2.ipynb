```python
from queue import Queue

graph={
    'A': ['B', 'C', 'E'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F', 'G'],
    'D': ['B'],
    'E': ['A', 'B', 'D'],
    'F': ['C'],
    'G': ['C']
}

visited={}
level={}
parent={}
bfs=[]
queue=Queue()

for node in graph.keys():
    visited[node]=False
    parent[node]=None
    level[node]=-1
    
s="A"
visited[s]=True
level[s]=0
queue.put(s)

while not queue.empty():
    u=queue.get()
    bfs.append(u)
    
    for v in graph[u]:
        if not visited[v]:
            visited[v]=True
            parent[v]=u
            level[v]=level[u]+1
            queue.put(v)

# Convert each node to uppercase and add it to the result list
result = [node.upper() for node in bfs]

print(result)

v="G"
path=[]
while v is not None:
    path.append(v)
    v=parent[v]
path.reverse()

# Convert each node in the path to uppercase and print it
print("Destination Sources")
for node in path:
    print(node.upper())
```

```python
graph = {
    'A': ['B', 'C', 'E'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F', 'G'],
    'D': ['B'],
    'E': ['A', 'B', 'D'],
    'F': ['C'],
    'G': ['C']
}

visited = {}
parent = {}
dfs = []


def dfs_util(node):
    visited[node] = True
    dfs.append(node)

    for neighbor in graph[node]:
        if not visited[neighbor]:
            parent[neighbor] = node
            dfs_util(neighbor)


for node in graph.keys():
    visited[node] = False
    parent[node] = None

start_node = "A"
dfs_util(start_node)

v = "G"
path = []

while v is not None:
    path.append(v)
    v = parent[v]

path.reverse()
print("DFS Traversal:")
print(dfs)
print("Destination Sources:")
print(path)
```
